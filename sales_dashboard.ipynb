### Using Python, Pandas, Plotly and Seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
sales_data = pd.read_excel(
    "US Superstore data.xls"
)

sales_data.head()
```

```python
sales_data = pd.read_excel(
    "US Superstore data.xls"
)

sales_data.head()
```

```python
sales_data.info()
```

```python
sales_data.isnull().sum()
```

```python
sales_data = sales_data.drop_duplicates()

sales_data.shape
```

```python
total_sales = sales_data['Sales'].sum()

total_profit = sales_data['Profit'].sum()

total_orders = sales_data['Order ID'].nunique()

print("Total Sales:", total_sales)

print("Total Profit:", total_profit)

print("Total Orders:", total_orders)
```

```python
region_sales = sales_data.groupby(
    'Region'
)['Sales'].sum().reset_index()

region_sales
```

```python
fig = px.bar(
    region_sales,
    x='Region',
    y='Sales',
    title='Revenue by Region',
    color='Region'
)

fig.show()
```

```python
fig = px.bar(
    region_sales,
    x='Region',
    y='Sales',
    title='Revenue by Region',
    color='Region'
)

fig.show()
```

```python
top_products = sales_data.groupby(
    'Product Name'
)['Sales'].sum().sort_values(
    ascending=False
).head(10)

top_products
```

```python
fig = px.bar(
    x=top_products.values,
    y=top_products.index,
    orientation='h',
    title='Top 10 Products by Sales'
)

fig.show()
```

```python
fig = px.bar(
    x=top_products.values,
    y=top_products.index,
    orientation='h',
    title='Top 10 Products by Sales'
)

fig.show()
```

```python
sales_data['Order Date'] = pd.to_datetime(
    sales_data['Order Date']
)

sales_data['Month'] = sales_data[
    'Order Date'
].dt.month

monthly_sales = sales_data.groupby(
    'Month'
)['Sales'].sum().reset_index()

monthly_sales
```

```python
fig = px.line(
    monthly_sales,
    x='Month',
    y='Sales',
    title='Monthly Sales Trend',
    markers=True
)

fig.show()
```
